## data analysis

start with looking into calender df

```python
calender_data.info()
```

```python
# we need to change the date column into date format 
calender_data['date'] = pd.to_datetime(calender_data.date)
calender_data.head()
```

```python
# create 3 new columns month, year, day of the week 
calender_data['month'], calender_data['year'] = calender_data.date.dt.month, calender_data.date.dt.year 
calender_data.head()
```

```python
from datetime import date

calender_data['day_of_week'] =  calender_data.date.dt.day_name()
calender_data.head()
```

```python
calender_data['weekend'] =  calender_data.day_of_week.apply(lambda x:x in(['Saturday', 'Sunday']))
calender_data.tail()
```
